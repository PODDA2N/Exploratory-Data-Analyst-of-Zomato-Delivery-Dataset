## Analyze the average Time_taken for orders to be delivered across different conditions (e.g., weather, traffic)

```python
delivery_time_taken = data.groupby (["Road_traffic_density","Weather_conditions"])["Time_taken (min)"].mean().reset_index()
delivery_time_taken
```

```python
plt.figure (figsize = (10,3))
sns.barplot(data = delivery_time_taken, x = "Road_traffic_density", y = "Time_taken (min)", hue = "Weather_conditions")
plt.title ("delivery_time_taken")
plt.show()
```

## What is the distribution of delivery_person_ratings across different city.

```python
average_ratings = data.groupby ("City")["Delivery_person_Ratings"].mean().reset_index().sort_values("Delivery_person_Ratings",ascending = False)
average_ratings
```

```python
plt.figure(figsize = (10,3))
sns.barplot(data = average_ratings, y = "Delivery_person_Ratings", x = "City")
plt.title("average_ratings")
plt.show()
```

## Calculate the average Time_taken per Delivery_person_ID to identify the fastest delivery person.

```python
fastest_delivery_person = data.groupby("Delivery_person_ID")["Time_taken (min)"].mean().reset_index().sort_values("Time_taken (min)",ascending = False).iloc[0]
fastest_delivery_person
```

## Check if Delivery_person_Age correlates with Time_taken to see if age impacts delivery times.

```python
sns.heatmap(data[["Delivery_person_Age","Time_taken (min)"]].corr(),annot = True)
```

## Group by City and analyze the average Time_taken to identify cities with longer delivery times.

```python
average_delivery_time = data.groupby ("City") ["Time_taken (min)"].mean().reset_index().sort_values("Time_taken (min)",ascending = False).iloc[0]
average_delivery_time
```

## Compare Time_taken under different Weather_conditions.

## Examine Road_traffic_density and its effect on Time_taken.

```python
plt.figure(figsize = (15,5))
sns.kdeplot(data = data, x = "Time_taken (min)", hue = "Road_traffic_density")
plt.title("Road_traffic_density and its effect on Time_taken")
plt.show()
```

## Compare Time_taken during Festival days versus non-festival days to see if festivals impact delivery times.

```python
average_delivery_time_festive_days = data.groupby("Festival")["Time_taken (min)"].mean().reset_index()
average_delivery_time_festive_days
```

```python
plt.figure(figsize = (10,3))
sns.barplot(data = average_delivery_time_festive_days, x = "Festival", y = "Time_taken (min)")
plt.title("Time_taken during Festival days")
plt.xlabel("Festival (Yes/No)")
plt.ylabel("Average Time Taken (min)")
plt.show()
```

```python
plt.figure(figsize = (15,5))
sns.boxplot(data = data, y = "Time_taken (min)", hue = "Festival")
plt.title("Time_taken during Festival days")
plt.show()
```

## Analyze the Type_of_order column to determine which types of food orders are most frequent.

```python
order_counts = data["Type_of_order"].value_counts().reset_index()
order_counts
```

```python
plt.figure(figsize = (10,3))
sns.countplot(data = order_counts, x = "Type_of_order")
plt.title("Frequency of Each Type of Order")
plt.show()
```

```python
proportion_order_counts = data["Type_of_order"].value_counts()
proportion_order_counts
```

```python
plt.figure(figsize = (5,5))
proportion_order_counts.plot.pie(labels = proportion_order_counts.index,autopct = "%1.2f%%",startangle = 90, wedgeprops = {"edgecolor" : "Red"}, explode = (0,0.05,0,0.1))
plt.title("Proportion of Each Type of Order")
plt.show()
```

## Analyze Type_of_vehicle to see if certain vehicles perform better in terms of delivery speed.
